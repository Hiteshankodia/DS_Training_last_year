```python
# replacing comma (,) with blank ''

df1["Total"] = df1["Total"].str.replace(',' , '')
df1
```

```python
# to make object data into numerics we use to_numeric
df1["Total"] = pd.to_numeric(df1["Total"])
```

```python
df1.dtypes
```

```python
df1
```

```python
x = df1["Units"].mean()
x = round(x)
units1 = df1["Units"].fillna(x)
units1
df1
```

```python
y = df1["Units"].median()
Units2 = df1["Units"].fillna(y)
Units2
```

```python
z = df1["Units"].mode()[0]
units3 = df1["Units"].fillna(z, inplace = True)
units3
df1
```

```python
# Dropping Duplicate Values
df1.drop_duplicates(inplace = True)
df1
```

```python
# Filtering Data

df2 = df1
for i in df2.index:
    if df2.loc[i, "Total"] > 500:
        df2.drop(i, inplace = True)
df2
```

```python
# Correlation - relationship between two columns (Features) - numerical

df2.corr()
```
